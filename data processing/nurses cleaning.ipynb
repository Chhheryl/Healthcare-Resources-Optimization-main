```python
import pandas as pd
import numpy as np
import os
```

```python
datadir = '../data'

file_path = os.path.join(datadir, 'nurses_deaggregated_by_population.xlsx')

# Read the Excel file into a pandas DataFrame
nurses_df = pd.read_excel(file_path)
```

```python
nurses_df.head()
```

```python
# Replace '**' with NaN in the 'tot_emp' column
nurses_df['tot_emp'] = nurses_df['tot_emp'].replace('**', np.nan)

# Drop rows where 'tot_emp' column contains NaN values
nurses_df = nurses_df.dropna(subset=['tot_emp'])
```

```python
nurses_df = nurses_df.reset_index(drop=True)
```

```python
nurses_df.columns
```

```python
nurses_df.drop()
```

```python
output_file = '../data/cleaned_nurses_data.xlsx'  

# Save the cleaned DataFrame to a new Excel file
nurses_df.to_excel(output_file, index=False)
```
